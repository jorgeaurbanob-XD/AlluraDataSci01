### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"Ingreso total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
categorias_tienda1 = tienda['Categoría del Producto'].value_counts()
print("--- Categorías más populares - Tienda 1 ---")
print(categorias_tienda1, "\n")

categorias_tienda2 = tienda2['Categoría del Producto'].value_counts()
print("--- Categorías más populares - Tienda 2 ---")
print(categorias_tienda2, "\n")

categorias_tienda3 = tienda3['Categoría del Producto'].value_counts()
print("--- Categorías más populares - Tienda 3 ---")
print(categorias_tienda3, "\n")

categorias_tienda4 = tienda4['Categoría del Producto'].value_counts()
print("--- Categorías más populares - Tienda 4 ---")
print(categorias_tienda4, "\n")
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda1 = tienda['Calificación'].mean()
promedio_calificacion_tienda2 = tienda2['Calificación'].mean()
promedio_calificacion_tienda3 = tienda3['Calificación'].mean()
promedio_calificacion_tienda4 = tienda4['Calificación'].mean()
```

```python
print(f"Calificación promedio Tienda 1: {promedio_calificacion_tienda1:.2f} / 5.00")
print(f"Calificación promedio Tienda 2: {promedio_calificacion_tienda2:.2f} / 5.00")
print(f"Calificación promedio Tienda 3: {promedio_calificacion_tienda3:.2f} / 5.00")
print(f"Calificación promedio Tienda 4: {promedio_calificacion_tienda4:.2f} / 5.00")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de tiendas y sus nombres para iterar fácilmente
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Configurar el tamaño general de los gráficos (4 filas x 2 columnas)
fig, axes = plt.subplots(4, 2, figsize=(16, 20))

for i, (df, nombre) in enumerate(zip(tiendas, nombres_tiendas)):
    # Contar la cantidad de veces que se vendió cada producto
    ventas_productos = df['Producto'].value_counts()

    # Extraer el Top 5 más vendidos y el Top 5 menos vendidos
    top_5 = ventas_productos.head(5)
    # Usamos sort_values para que el gráfico de los menos vendidos se vea ordenado de menor a mayor
    bottom_5 = ventas_productos.tail(5).sort_values(ascending=True)

    # --- Impresión en texto ---
    print(f"--- {nombre} ---")
    print("Top 5 Más Vendidos:")
    print(top_5.to_string())
    print("\nTop 5 Menos Vendidos:")
    print(bottom_5.to_string())
    print("\n" + "="*30 + "\n")

    # --- Gráfico Top 5 Más Vendidos (Columna Izquierda) ---
    sns.barplot(x=top_5.values, y=top_5.index, ax=axes[i, 0], palette='viridis')
    axes[i, 0].set_title(f'Top 5 MÁS Vendidos - {nombre}')
    axes[i, 0].set_xlabel('Cantidad Vendida')
    axes[i, 0].set_ylabel('Producto')

    # --- Gráfico Top 5 Menos Vendidos (Columna Derecha) ---
    sns.barplot(x=bottom_5.values, y=bottom_5.index, ax=axes[i, 1], palette='magma')
    axes[i, 1].set_title(f'Top 5 MENOS Vendidos - {nombre}')
    axes[i, 1].set_xlabel('Cantidad Vendida')
    axes[i, 1].set_ylabel('')

# Ajustar el espacio entre los gráficos para que se vea ordenado
plt.tight_layout()
# Mostrar los gráficos
plt.show()
```

# 5. Envío promedio por tienda

```python
promedio_envio_tienda1 = tienda['Costo de envío'].mean()
promedio_envio_tienda2 = tienda2['Costo de envío'].mean()
promedio_envio_tienda3 = tienda3['Costo de envío'].mean()
promedio_envio_tienda4 = tienda4['Costo de envío'].mean()

# Imprimir los resultados con formato de moneda
print("--- Costo de Envío Promedio por Tienda ---")
print(f"Tienda 1: ${promedio_envio_tienda1:,.2f}")
print(f"Tienda 2: ${promedio_envio_tienda2:,.2f}")
print(f"Tienda 3: ${promedio_envio_tienda3:,.2f}")
print(f"Tienda 4: ${promedio_envio_tienda4:,.2f}")
```

```python
# Definimos las listas para iterar (asumiendo que ya tienes tienda, tienda2, tienda3 y tienda4 cargadas)
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# 1. Calculamos las métricas para cada tienda usando comprensión de listas
ingresos_totales = [t['Precio'].sum() for t in tiendas]
calificacion_promedio = [t['Calificación'].mean() for t in tiendas]
envio_promedio = [t['Costo de envío'].mean() for t in tiendas]

# 2. Configuramos el lienzo para 3 subgráficos en una sola fila
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.set_theme(style="whitegrid")

# --- Gráfico 1: Ingresos Totales ---
sns.barplot(x=nombres_tiendas, y=ingresos_totales, ax=axes[0], palette="Greens_d")
axes[0].set_title('Facturación Total por Tienda', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Ingresos ($)')
# Añadir los valores sobre las barras
for i, v in enumerate(ingresos_totales):
    axes[0].text(i, v + (max(ingresos_totales)*0.01), f'${v:,.0f}', ha='center', fontsize=11)

# --- Gráfico 2: Calificación Promedio ---
sns.barplot(x=nombres_tiendas, y=calificacion_promedio, ax=axes[1], palette="Blues_d")
axes[1].set_title('Calificación Promedio de Clientes', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Estrellas (0 a 5)')
axes[1].set_ylim(0, 5) # Fijamos el límite de 0 a 5 estrellas
# Añadir los valores sobre las barras
for i, v in enumerate(calificacion_promedio):
    axes[1].text(i, v + 0.1, f'{v:.2f} ★', ha='center', fontsize=11)

# --- Gráfico 3: Costo de Envío Promedio ---
sns.barplot(x=nombres_tiendas, y=envio_promedio, ax=axes[2], palette="Oranges_d")
axes[2].set_title('Costo Promedio de Envío', fontsize=14, fontweight='bold')
axes[2].set_ylabel('Costo ($)')
# Añadir los valores sobre las barras
for i, v in enumerate(envio_promedio):
    axes[2].text(i, v + (max(envio_promedio)*0.01), f'${v:,.0f}', ha='center', fontsize=11)

# Ajustar el diseño para que todo encaje perfecto
plt.tight_layout()
plt.show()
```
